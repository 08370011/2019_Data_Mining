## 將資料處理成 tidy data

### 1. 使用的資料介紹
- [Kaggle Titanic Dataset](https://www.kaggle.com/c/titanic)

- train.csv：(891, 12)
- test.csv：(418, 11)

|欄位|columns|
|:---:|:---:|
|PassengerId|乘客ID編號|
|Pclass|船票等級|
|Name|姓名|
|Sex|性別|
|Age|年齡|
|SibSp(Siblings + Spouses)|兄弟姊妹＋老婆丈夫數量|
|Parch(Parents + Children)|父母小孩的數量|
|Fare|票的費用|
|Ticket|票的號碼|
|Cabin|船艙號碼|
|Embarked|出發港口|
|Survived|是否存活|

### 3. 結果呈現

```python
# import package
import numpy as np
import pandas as pd
```

```python
# read the file
train = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
```

```python
# view the data
train.head()
```

```python
# shape of train data, test data
print(f'train shape : {train.shape}')
print(f'test shape : {test.shape}')
```

```python
# describe of train data
train.describe()
```

```python
# describe of test data
test.describe()
```

```python
# concat train and test
df = pd.concat([train, test], join='inner')
df.head()
```

```python
df.tail()
```

```python
print(f'df shape : {df.shape}')
```
