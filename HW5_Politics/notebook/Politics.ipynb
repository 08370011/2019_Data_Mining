### Drop/Add Columns

#### Company
1. drop不需要的columns：統一編號、P、序號、支出金額
2. 當選註記、金錢類：0, 1
3. 交易日期：年月、年、月、日
4. 推薦政黨：民主進步黨、中國國民黨、其他

```python
company = company.drop(['統一編號', 'P', '序號', '支出金額'], axis=1)
company['當選註記'] = company['當選註記'].fillna('否')
company[company['當選註記']=='*'] = company[company['當選註記']=='*'].replace('*', '是')
company = company.fillna('其他')
company['當選註記'] = company['當選註記'].apply(lambda x:1 if x=='是' else 0)
company['金錢類'] = company['金錢類'].apply(lambda x:1 if x=='是' else 0)
```

```python
company['年月'] = company['交易日期'].apply(lambda x:x.split('/')[0]+ '-' +x.split('/')[1])
company['年'] = company['交易日期'].apply(lambda x:x.split('/')[0]).astype(int)
company['月'] = company['交易日期'].apply(lambda x:x.split('/')[1]).astype(int)
company['日'] = company['交易日期'].apply(lambda x:x.split('/')[2])
company = company.replace({'日': {'1  I': '11'}})
company['日'] = company['日'].astype(int)
company = company.drop(['交易日期'], axis=1)
company.replace({'104-7':'104-07','104-8':'104-08',
                '104-9':'104-09','105-1':'105-01'}, inplace=True)
```

```python
company['政黨'] = company['推薦政黨'].map({'民主進步黨':'民主進步黨', '中國國民黨':'中國國民黨'})
company['政黨'] = company['政黨'].fillna('其他')
```

#### Basic
1. drop不需要的columns：超過三萬元之收入
2. 當選註記：0, 1
3. 交易日期：年月、年、月、日
4. 縣市
5. 性別：0, 1
6. 年齡
7. 推薦政黨：民主進步黨、中國國民黨、其他

```python
basic = basic.drop(['超過三萬元之收入'], axis=1)
basic['當選註記'] = basic['當選註記'].fillna('否')
basic[basic['當選註記']=='*'] = basic[basic['當選註記']=='*'].replace('*', '是')
basic = basic.fillna('其他')
basic.insert(loc=0, column='縣市', value=basic['地區'].apply(lambda x:x[0:3]))
basic = basic.replace({'縣市': {'山原原': '山原原住民', '平原原': '平原原住民'}})
```

```python
def get_float(col):
    basic[col] = basic[col].apply(lambda x:x.replace('%', ''))
    basic[col] = basic[col].astype(float)/100
    
def get_int(col):
    basic[col] = basic[col].apply(lambda x:x.replace(',', ''))
    basic[col] = basic[col].astype(int)
```

```python
col = ['得票率', '個人捐贈比例', '營利事業捐贈比例', '政黨捐贈收入比例', '人民團體收入比例', '匿名捐贈比例', '其他收入比例']
for i in col:
    get_float(i)
    
col = ['個人捐贈收入', '人民團體捐贈收入', '匿名捐贈收入', '總支出', '其他收入', '政黨捐贈收入']
for i in col:
    get_int(i)    
    
col = ['當選註記', '是否現任']
for i in col:
    basic[i] = basic[i].apply(lambda x:1 if x=='是' else 0)    
```

```python
basic['性別'] = basic['性別'].apply(lambda x:1 if x=='男' else 0)
basic['年齡'] = 2019 - basic['出生年次']
basic['政黨'] = basic['推薦政黨'].map({'民主進步黨':'民主進步黨', '中國國民黨':'中國國民黨'})
basic['政黨'] = basic['政黨'].fillna('其他')
```

### Missing Value

```python
def na_check(df):
    missing = pd.DataFrame(df.isnull().sum(), columns=['missing'])
    missing = missing.sort_values(ascending=False, by=['missing'])
    missing = missing[missing['missing']>0]
    if len(missing)==0:
        print('No Missing Value !')
    else:
        display(missing)
```

### 各黨派捐贈比例分佈
1. 這邊可以看出三種黨派的政治獻金組成都不太一樣
2. 國民黨的候選人政治獻金的組成以政黨捐贈為主
3. 民進黨是以個人+營利事業捐贈為主、政黨捐贈不到5%
4. 其他黨派則是以個人捐贈為主，再來才是政黨捐贈

```python
df = basic.groupby('政黨').mean()[['個人捐贈比例', '營利事業捐贈比例', '政黨捐贈收入比例', '人民團體收入比例','匿名捐贈比例', '其他收入比例']].transpose()
df0 = pd.DataFrame({'比例種類':list(df.index), '比例':df['中國國民黨'].values, '政黨':'中國國民黨'})
df1 = pd.DataFrame({'比例種類':list(df.index), '比例':df['民主進步黨'].values, '政黨':'民主進步黨'})
df2 = pd.DataFrame({'比例種類':list(df.index), '比例':df['其他'].values, '政黨':'其他'})
df = pd.concat([df0, df1, df2], axis=0)
```

```python
plt.figure(figsize=(17,5))
sns.barplot(x=df.比例種類, y=df.比例, hue=df.政黨)
```

### 當選、是否現任比較
#### 年紀
- 可以看出現任的立法委員的年紀偏大，大多都集中在5, 60歲
- 而這次當選的立法委員的年紀有**年輕化的趨勢**
- 變成集中在4, 50歲

```python
on_ = basic[['姓名', '年齡', '當選註記', '是否現任']]
on_['年齡_區間']= (on_['年齡']/5).astype(int)*5
df = on_.groupby('年齡_區間').sum()
df1 = pd.DataFrame({'年齡_區間':list(df.index), '人數':df['當選註記'], '類型':'當選註記'})
df2 = pd.DataFrame({'年齡_區間':list(df.index), '人數':df['是否現任'], '類型':'是否現任'})
df = pd.concat([df1, df2], axis=0)
```

```python
plt.figure(figsize=(17,5))
sns.set(font_scale=1.5)
sns.set_style("whitegrid",{"font.sans-serif":['Microsoft JhengHei']})
sns.barplot(x=df.年齡_區間, y=df.人數, hue=df.類型)
```
